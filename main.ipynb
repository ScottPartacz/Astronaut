```python
import csv
import random

class astronaut():
    
    def __init__(self,name,status,flighthours,year = None,group = None,birthdate = None,birthplace = None,gender = None,almamater = None,undergrad = None,grad = None,rank = None,branch = None,flights = None,spacewalks = None,spacewalkhour = None,missions = None,deathdate = None,deathmission = None ):
        self._name = name
        self._year = year
        self._group = group
        self._status = status
        self._birthdate = birthdate
        self._birthplace = birthplace
        self._gender = gender
        self._almamater = almamater
        self._undergrad = undergrad
        self._grad = grad
        self._rank = rank
        self._branch = branch
        self._flights = flights
        self._flighthours = flighthours
        self._spacewalks = spacewalks
        self._spacewalkhours = spacewalkhour
        self._missions = missions
        self._deathdate = deathdate
        self._deathmission = deathmission
    
    def __eq__(self,other):
        if (self._flighthours == other._flighthours):
            return self._flighthours == other._flighthours
        else:
            return False

    def __gt__(self,other):
        if (self._flighthours > other._flighthours):
            return self._flighthours > other._flighthours
        else:
            return False
    
    def __ge__(self,other):
        if (self._flighthours >= other._flighthours):
            return self._flighthours >= other._flighthours
        else:
            return False
    
    def __str__(self):
        return (self._name + "," + self._status)
    


with open("astronauts.csv") as file:

    data = [{key: value for key,value in row.items()} for row in csv.DictReader(file)]

astronauts = []


for elements in data:

    temp = astronaut(elements["Name"],elements["Status"],elements["Space Flight (hr)"],elements["Year"],elements["Group"],elements["Birth Date"],elements["Birth Place"],elements["Gender"],elements["Undergraduate Major"],elements["Graduate Major"],elements["Military Rank"],elements["Military Branch"],elements["Space Flights"],elements["Space Walks"],elements["Space Walks (hr)"],elements["Missions"],elements["Death Date"],elements["Death Mission"])
    
    astronauts.append(temp)

print(astronauts[0].__dict__.values())

print(random.choice(astronauts) == random.choice(astronauts))

print(random.choice(astronauts) > random.choice(astronauts))

print(random.choice(astronauts) >= random.choice(astronauts))

[print(x) for x in astronauts]

```
